```python
from autogen_ext.models.ollama import OllamaChatCompletionClient
from autogen_core.models import ModelInfo, ModelFamily

model_information = ModelInfo(
    vision=False,
    function_calling=False,
    json_output=False,
    structured_output=False,
    multiple_system_messages=False,
    family=ModelFamily.ANY
)

ollama_client = OllamaChatCompletionClient(model="gemma3:1b", model_info=model_information)
```

```python
from autogen_agentchat.agents import AssistantAgent
from autogen_agentchat.teams import DiGraphBuilder, GraphFlow
import asyncio

agent_a = AssistantAgent("analyzer", model_client=ollama_client, system_message="You create a text")
agent_b = AssistantAgent("validator", model_client=ollama_client, system_message="You review the text and correct mistakes.")
agent_c = AssistantAgent("formatter", model_client=ollama_client, system_message="You format the final output as a report.")

builder = DiGraphBuilder()
builder.add_node(agent_a).add_node(agent_b).add_node(agent_c)
builder.add_edge(agent_a, agent_b)
builder.add_edge(agent_b, agent_c)
graph = builder.build()

flow = GraphFlow(participants=[agent_a, agent_b, agent_c], graph=graph)

async def main():
    async for event in flow.run_stream(task="Create an history and review it"):
        print(event)

await main()
```
